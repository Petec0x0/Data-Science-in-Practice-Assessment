```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

from scipy.stats import pearsonr, spearmanr
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

<h2 id="understanding_data">Understanding the Data</h2>

### `ConcreteCompressiveStrength.csv`:
<pre>
    <b>
    Task:
        Predict the compressive strenght of concrete based on its composition and its age

    Dataset Description:
    The dataset has the following input features
        Cement: kg per m^3
        Blast_Furnace_Slag: kg per m^3
        Fly_Ash: kg per m^3
        Water: kg per m^3
        Superplasticizer: kg per m^3
        Coarse_Aggregate: kg per m^3
        Fine_Aggregate: kg per m^3
        Age: in days

    The output feature is
        Concrete_compressive_strength: MPa
     </b>   
<pre>

<h2 id="reading_data">Reading the data in</h2>

```python
data = pd.read_csv("ConcreteCompressiveStrength.csv")

# take a look at the dataset
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.hist(figsize=(20,15))
plt.show()
```

__Checking if there is any linear relaionship b/w `Concrete_compressive_strength` and the independent variables__

```python
plt.figure(figsize=(20,15))
plt.scatter(data[['Cement']], data[['Concrete_compressive_strength']], color='b')
plt.scatter(data[['Blast_Furnace_Slag']], data[['Concrete_compressive_strength']], color='g')
plt.scatter(data[['Fly_Ash']], data[['Concrete_compressive_strength']], color='r')
plt.scatter(data[['Water']], data[['Concrete_compressive_strength']], color='c')
plt.scatter(data[['Superplasticizer']], data[['Concrete_compressive_strength']], color='m')
plt.scatter(data[['Coarse_Aggregate']], data[['Concrete_compressive_strength']], color='y')
plt.scatter(data[['Fine_Aggregate']], data[['Concrete_compressive_strength']], color='k')
plt.legend(['Cement','Blast_Furnace_Slag','Fly_Ash','Water','Superplasticizer','Coarse_Aggregate','Fine_Aggregate'])
plt.ylabel('Concrete_compressive_strength')
plt.show()
```

__Checking if there is any linear relaionship b/w `Concrete_compressive_strength` and other variables using Spearmans correlation coeficient__

```python
corr, _ = spearmanr(data[['Cement']], data[['Concrete_compressive_strength']])
print('Spearmans correlation: %.3f' % corr)
```

__Data Spliting__

```python
features = data.drop(['Concrete_compressive_strength'], axis=1)
labels = data[['Concrete_compressive_strength']]
train_x, test_x, train_y, test_y = train_test_split(features,labels, test_size=0.2, random_state=3)
```

```python
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
stdScaler = StandardScaler()
stdScaler.fit(train_x)

train_x = stdScaler.transform(train_x)
test_x = stdScaler.transform(test_x)
```

```python
poly = PolynomialFeatures(degree=3)
poly_train_x = poly.fit_transform(train_x)
regr = linear_model.LinearRegression()
regr.fit(poly_train_x, train_y)

print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
XX = np.arange(0.0, 10.0, 0.1)
yy = regr.intercept_[0]+ regr.coef_[0][1]*XX+ regr.coef_[0][2]*np.power(XX, 2)
plt.figure(figsize=(20,15))
plt.plot(XX, yy, '-r' )
plt.scatter(data[['Cement']], data[['Concrete_compressive_strength']], color='b')
plt.scatter(data[['Blast_Furnace_Slag']], data[['Concrete_compressive_strength']], color='g')
plt.scatter(data[['Fly_Ash']], data[['Concrete_compressive_strength']], color='r')
plt.scatter(data[['Water']], data[['Concrete_compressive_strength']], color='c')
plt.scatter(data[['Superplasticizer']], data[['Concrete_compressive_strength']], color='m')
plt.scatter(data[['Coarse_Aggregate']], data[['Concrete_compressive_strength']], color='y')
plt.scatter(data[['Fine_Aggregate']], data[['Concrete_compressive_strength']], color='k')
plt.legend(['Cement','Blast_Furnace_Slag','Fly_Ash','Water','Superplasticizer','Coarse_Aggregate','Fine_Aggregate'])
plt.ylabel('Concrete_compressive_strength')
plt.show()
```

```python
train_x_poly = poly.fit_transform(train_x)
train_y_hat = regr.predict(train_x_poly)

test_x_poly = poly.fit_transform(test_x)
test_y_hat = regr.predict(test_x_poly)


print("R2-score for traning set: %.2f" % r2_score(y_pred=train_y_hat , y_true=train_y) )
print("R2-score for test set: %.2f" % r2_score(y_pred=test_y_hat , y_true=test_y) )
```
